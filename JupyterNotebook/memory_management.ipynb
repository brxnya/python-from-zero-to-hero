## В принципе, один объект может быть "помечен" множеством меток

* Есть две стратегии сборки мусора:
    * reference counting
    * tracing
* Подсчёт ссылок не работает на циклических ссылках
(у двух объектов ссылающихся друг на друга счётчик не становится ниже 1)
* tracing - построение графа объектов, отслеживание поколений объектов
* иногда запускается GC (garbage collector), который маркирует объекты и удаляет "осиратевшие"
* GC отслеживает длительность жизни объектов и группирует их по трём поколениям:
    * 0(новые), 1, 2(старые)
* когда достигается некий порог потребления памяти в поколении, на нём запускается GC и на тех, которые младше
* если объект пережил сборку мусора - он переходит в старшее поколение
